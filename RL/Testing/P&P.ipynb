```python
def factores_primos(x):
    factores = []
    divisor = 2
    while x > 1:
        if x % divisor == 0:
            factores.append(divisor)
            x //= divisor
        else:
            divisor += 1
    return factores

recompensas = [18,8,16,12]
hora = range(7,21,1)
arms = [0,0,0,0]
contadores = [0,0,0,0]

for _ in hora:
    numero = len(factores_primos(_))
    camino_idx = numero - 1
    
    if _%4 > 0:  # Puente abierto
        arms[camino_idx] += recompensas[camino_idx]
        contadores[camino_idx] += 1
    else:  # Puente cerrado
        arms[camino_idx] += 0
        contadores[camino_idx] += 1
arms = [arms[i]/contadores[i] if contadores[i] > 0 else 0 for i in range(4)]
arms = [round(x, 2) for x in arms]

arms
```
